```python
import os
import numpy as np
import imageio.v3 as iio
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
index = 100
png_path = os.listdir("labeled_images")[index]
json_path = os.listdir("labels")[index]

im = iio.imread("labeled_images/"+png_path)
plt.imshow(im)
```

```python
#takes a single json file and turns it into a png, putting the new png label and the original image into the directory after the "-o" part below
def labelme_json_to_dataset(json_path):
    os.system("labelme_json_to_dataset "+json_path+" -o temp")
```

```python
#loops through every file in a directory that contains just the json labels, then turns those into pngs using the labelme_json_to_dataset python file.
#then these are saved to the directory specified in that command, and then moved from that temporary folder to a permanent folder
#so that we only get the label.

json_label_folder_path = "json_labels"

i = 1
for json_path in os.listdir(json_label_folder_path):
    labelme_json_to_dataset(json_label_folder_path+"/"+json_path)
    os.replace("temp/label.png","png_labels/"+json_path[:-5]+".png")
    print("Completed image", i, end="\r")
    i += 1
print("")
```
